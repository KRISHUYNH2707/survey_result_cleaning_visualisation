```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_data = pd.read_excel('./extracted_data.xlsx',header=1)
```

```python
raw_data
```

```python
raw_data['Percentage'] = raw_data['Percentage'].replace('%','').astype(float)
```

```python
raw_data
```

```python
raw_data.fillna(method='ffill', inplace=True)
raw_data
```

```python
raw_data['Percentage'].fillna(0, inplace=True)
raw_data['Level'].fillna('', inplace=True)
```

```python
new_df = pd.DataFrame(columns=['Strongly disagree', 'Disagree',
                  'Unsure', 'Agree', 'Strongly agree'])
questions = raw_data['Variable'].unique()
new_df['Questions'] = questions
new_df.set_index('Questions', inplace=True)
```

```python
raw_data
```

```python
new_df
```

```python
levels = new_df.columns
for question in questions:
    for level in levels:
        percent = (raw_data.loc[(raw_data['Variable'] == question) & (raw_data['Level'] == level), 'Percentage'])
        print(percent)
        if percent.empty:
            print('empty')
            percent = 0
        else:
            percent = percent.values[0]
        print(question, level, percent)
        print(f'new_df.loc[{question}, {level}] = {percent}')
        new_df.loc[question, level] = percent
```

```python
empty_rows = ['Burden factor', 'Appropriate factor', 'Concern about stopping factor', 'Involvement factor', 'Global Questions']

for row in empty_rows:
    new_df.loc[row] = 0
```

```python
new_df
```

```python
labels = new_df.index
colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, new_df.shape[1]))
categories = new_df.columns
```

```python
for label in labels:
    print(label)
```

```python
new_df
```

```python
# Change the data in each row but still keeps everything 100
```

```python
fig, ax = plt.subplots(figsize=(20, 30))
ax.xaxis.grid(True, linestyle='--', alpha=1, zorder=0)
ax.invert_yaxis()
# Remove the plot frame lines
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
# Remove x and y axis ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Use a set to keep track of categories that have been added to the legend
added_categories = set()

for label in labels:
    cum_sum = 0
    for i, (color, category) in enumerate(zip(colors, categories)):
        bar_width = new_df.loc[label, category]
        if category not in added_categories:
            rects = ax.barh(label, bar_width, left=cum_sum, height=0.5,
                    label=category, color=color, zorder=3)
            added_categories.add(category)
        else:
            rects = ax.barh(label, bar_width, left=cum_sum, height=0.5,
                    color=color, zorder=3)
        # r, g, b, _ = color
        # text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        # if bar_width > 0:
        #     ax.bar_label(rects, label_type='center', color=text_color, fontsize=20)
        cum_sum += bar_width

# Set legend at the bottom of the plot
ax.legend(loc='upper center', bbox_to_anchor=(0.1, -0.05),
          fancybox=False, shadow=True, ncol=5, prop={'size': 20})


# add y gridlines behind the bars
# plt.grid(axis='x', linestyle='--', alpha=0.5, zorder=0)

 
# Add chart title
ax.set_title('Survey Results', loc='left', fontsize=35)

# Add padding between chart title and the plot
plt.subplots_adjust(top=0.85)

# Set font size of text in y axis
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

# format x axis as percentage
ax.xaxis.set_major_formatter(plt.FuncFormatter('{:.0f}%'.format))

# save the plot in full size
plt.savefig('survey_results.png', bbox_inches='tight')
```
